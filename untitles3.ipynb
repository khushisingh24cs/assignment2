```python
q1=20.125
q2=28
q3=38
iqr=(q3-q1)
l=q1-(iqr*1.5)
u=q3+(iqr*1.5)
print(l)
print(u)
#formula for outlier
```

```python
outlier_age=(df["Age"]<l)|(df["Age"]>u)
#outlier defined
```

```python
print(df[outlier_age])
```
